```python
#First we create our variables and empty lists. Next we create another list with a series of ascending numbers 0-1499
#so that we can iterate through. We'll want to find a city for each number. For each number in that list, we'll choose and
#store a random latitude and longitude using .uniform(). We'll then run each number associated with its respective lats and lngs
#through our python library and use its functions to find the nearest city based on those inputs. We append that result to another empty list
#and restart the for loop. We'll add a try/except function in the case that we're not able to find the nearest city. We will find them all.
lats = 0
lngs = 0
lats_list = []
lngs_list = []
cities = []
country_codes = []
sample_list = []
for i in range(1500):
    sample_list.append(i)
#here we create a for loop to circle through the data, print out variables for each of our columns. 
#we next need to create a dataframe that holds these variables in order and vuala!
#we'll need to remove duplicate values from the df
for city in sample_list:
    try:
        lats = random.uniform(-90,90)
        lats_list.append(lats)
        lngs = random.uniform(-180,180)
        lngs_list.append(lngs)
        city = citipy.nearest_city(lats, lngs)
        cities.append(city.city_name)
        country_codes.append(city.country_code)
        print(f"{city.city_name}, {city.country_code} found! Appending to cities.")
    except:
        print("Error. Moving on.")
        pass
```

```python
#we create many empty lists to append into as we collect our data from our OpenWeather
#add a try/except function to skip over cities of which OpenWeather has no data
#I've selected metric for our units. We'll use an f string to make sure the city name is printed as the function searches. 
#we use our cities list filled with names to run through a request.get and then append the appropriate variables into many other lists.


max_temps = []
only_humids = []
only_cloudys = []
only_windspeeds = []
only_countrys = []
only_dates = []
units = "metric"
query_url = f"{url}&appid=f26d926a5842d291406598018c829aaf&units={units}&q="

for x in cities:
    try:
        response = requests.get(query_url + x).json()
        max_temps.append(response["main"]["temp"])
        only_humids.append(response["main"]["humidity"])
        only_cloudys.append(response["clouds"]["all"])
        only_windspeeds.append(response["wind"]["speed"])
        only_countrys.append(response["sys"]["country"])
        only_dates.append(response["dt"])
        print(f"{x} found - continuing...")
#I'm using np.nan to fill the list in the appropriate spots so we can later use .dropna() to get rid of entire rows.
    except:
        max_temps.append(np.nan)
        only_humids.append(np.nan)
        only_cloudys.append(np.nan)
        only_windspeeds.append(np.nan)
        only_countrys.append(np.nan)
        only_dates.append(np.nan)
        print(f"no info available on {x} appending NaN - continuing...")
        pass
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#here we put all of our lists together to form a dataframe. 
d = {"city":cities, "latitude":lats_list, "longitude":lngs_list,"max temp":max_temps, "humidity":only_humids, 
     "cloudiness":only_cloudys, "wind speeds":only_windspeeds, "country":only_countrys, "date":only_dates}
df = pd.DataFrame(data=d)
df.to_csv('raw_weather_data')
df
```

```python
#we drop any row that has a NaN value and drop duplicates of the same subset ["city"], meaning the same name. 
df = df.dropna(how='any')
df = df.drop_duplicates(subset=["city"], keep="first")
df.to_csv('clean_all_weather_data')
```

## Linear Regression

```python
#First I need to split the categories that I'll be working with. I'll need two data frames - N. Hemi and S. Hemi
#we could use loc, but then we'd have to sanitize some of the series as not all are integers. This is easier. 
df_south = df[df["latitude"] < 0]
df_south.to_csv('df_south')
df_north = df[df["latitude"] >= 0]
df_north.to_csv('df_north')
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Our regression is a mess of code. Essentially we first find the correlation using st.pearsonr(x,y). Next we print the correlation and round. We use linregress() magic to define our statistics. 
#we'll create our regress values to plot our regression line against our data, manually write out our line, plot our line, plot our data. I choose not to do anything too fancy with plotting. 
#We do this 7 more times using different variables and different data. 
correlation = st.pearsonr(df_north["latitude"], df_north["max temp"])
print(f"The correlation between both factors is {round(correlation[0],2)}")
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(df_north["latitude"], df_north["max temp"])

regress_values = df_north["latitude"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
x0 = df_north["latitude"]
y0 = df_north["max temp"]
plt.plot(x0,regress_values,"--",color='r')
print(f"The r-squared is: {rvalue**2}")
plt.annotate(line_eq,(2,-5),fontsize=15,color="red")
print(line_eq)
plt.scatter(x0, y0)
plt.title("Northern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.ylabel("Max Temperature (C)")
plt.xlabel("Latitude")
plt.savefig("/Users/jlh/ucb-bel-data-pt-08-2020-u-c/Saved_Docs/API_HW")
plt.show()
```
